## Data Cleaning

**1.** Create a copy of base data for manupulation & processing

```python
new_previous_df = previous_df.copy()
```

```python
new_previous_df.head()
```

2. Remove columns not required for processing

```python
columns_to_drop = ['NFLAG_INSURED_ON_APPROVAL','SK_ID_PREV','SK_ID_CURR','FLAG_LAST_APPL_PER_CONTRACT','NFLAG_LAST_APPL_IN_DAY','NAME_TYPE_SUITE','SELLERPLACE_AREA','WEEKDAY_APPR_PROCESS_START','HOUR_APPR_PROCESS_START']

common_columns = list(set(columns_to_drop) & set(new_previous_df.columns))

new_previous_df.drop(columns=common_columns, axis=1, inplace=True)
new_previous_df.head()
```

**2.** Count of null values

```python
new_previous_df.isnull().sum()
```

**3.** Check for the percentage of missing values

Missing values for previous data

```python
percent_missing = new_previous_df.isnull().sum() * 100 / len(new_previous_df)
p_missing_value_df = pd.DataFrame({'column_name': new_previous_df.columns,
                                 'percent_missing': percent_missing})
```

**4. Missing Value Treatement**

**4a. Drop columns**

Drop the data for columns with more than 40% of null values. There wouldn't be any helpful info even if we choose to remain these columns. We can't replace it with other values like mean as the existing values available that we can use to calculate mean is less than 60%(2/3).


```python
columns_to_drop = ['AMT_DOWN_PAYMENT','RATE_DOWN_PAYMENT', 'RATE_INTEREST_PRIMARY', 'RATE_INTEREST_PRIVILEGED',
                   'DAYS_FIRST_DRAWING', 'DAYS_FIRST_DUE', 'DAYS_LAST_DUE_1ST_VERSION',
                   'DAYS_LAST_DUE', 'DAYS_TERMINATION']

common_columns = list(set(columns_to_drop) & set(new_previous_df.columns))

new_previous_df.drop(columns=common_columns, axis=1, inplace=True)
new_previous_df.head()
```

**4b. Replace values** 

Replace the null values for columns with null values between 20% to 40%. We'll fill the missing data with mean value of each category.

```python
columns_to_fill_mean = ['AMT_ANNUITY','AMT_GOODS_PRICE', 'CNT_PAYMENT']

for col in columns_to_fill_mean:
    new_previous_df[col].fillna(new_previous_df[col].mode()[0], inplace=True)

new_previous_df.info()
```

Since the % of the remaining columns with null values compared to total dataset is very low(less than 1%), it is safe to ignore them from further processing.

### Numerical Analysis

Boxplot for numerical columns as they contain too many outliers

```python
plt.figure(figsize=(10,12))

columns_to_plot=['AMT_ANNUITY','AMT_APPLICATION','AMT_CREDIT','AMT_GOODS_PRICE']

for i, predictor in enumerate(columns_to_plot):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=new_previous_df, y=predictor, orient = "v", color = "orange")

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
percentile_5 = new_previous_df['AMT_ANNUITY'].quantile(0.05)
percentile_95 = new_previous_df['AMT_ANNUITY'].quantile(0.95)

# Create a single figure for all the plots
plt.figure(figsize=(12, 8))

# Loop through the columns and plot KDE plots for filtered data
for i, predictor in enumerate(columns_to_plot):
    filtered_data = new_previous_df[(new_previous_df[predictor] >= percentile_5) &
                                    (new_previous_df[predictor] <= percentile_95)]
    
    plt.subplot(2, 2, i+1)  # Create a subplot
    sns.kdeplot(filtered_data[predictor], fill=True)
    plt.title(predictor)
    plt.xlabel(predictor)
    plt.ylabel('Density')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
plt.show()
```
